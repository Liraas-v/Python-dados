Calcular a média do estudante com até uma casa decimal.

Os dados recebidos correspondem a um dicionário com as chaves indicando o trimestre em questão e os valores das notas de cada trimestre do estudante em cada matéria

```python
notas = {"1 trimestre": 8.5, "2 trimestre": 9.5, "3 trimestre": 7}
notas
```

```python
soma = 0

for nota in notas.values():
    soma += nota

soma
```

```python
# usando a função embutida sum( built-in function)
soma_das_notas = sum(notas.values())
soma_das_notas
```

```python
# usando a função embutida len ( ver o comprimento)
qtd_notas = len(notas)
qtd_notas
```

```python
# calculando a média
media = soma_das_notas / qtd_notas
media
```

```python
# arredondar a média usando round():

media = soma_das_notas / qtd_notas
media = round(media,1)
media
```

```python
notas_estudantes = [ 10,9,6.7,8.5]

def media(lista):
    calculo = sum(lista) / len(lista)
    calculo = round(calculo,1)
    return calculo
```

```python
media(notas_estudantes)
```

```python
resultado = media(notas_estudantes)
resultado
```

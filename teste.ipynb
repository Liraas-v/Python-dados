```python
import matplotlib.pyplot as plt 
from random import choice # biblioteca usada para escolher um elemento aleatorio

estudantes = [ "João", " Maria", "José", " Matheus", " Rodrigo", "Julia", " Bruna ", " Carlos"]
notas =  [ 8.5, 9, 6.5, 7.3, 9, 10, 5, 4.5]

plt.bar(x = estudantes, height = notas)

estudantes_2 = [" João", "Maria", " José", " Ana"]
estudante = choice(estudantes_2)
estudante

listas = [ 1,2,3,4,5,6,7,8,9]
lista = choice(listas)
lista
```
